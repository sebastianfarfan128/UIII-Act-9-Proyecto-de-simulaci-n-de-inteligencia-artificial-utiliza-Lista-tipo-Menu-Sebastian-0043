### Proyecto de simulación de inteligencia artificial  utiliza Lista,con if, else, elif  while tipo Menu
### 18. Administrador

```python
# Inicialización del sistema
usuarios = [] # La lista comienza vacía [4]
sistema_activo = True
limite_capacidad = 5

print("--- IA SISTEMA ADMINISTRADOR DE ACCESOS v1.0 ---")

# Bucle principal interactivo
while sistema_activo:
    # Lógica de "Percepción e Inteligencia" de la IA
    # La IA evalúa la situación basándose en la longitud de la lista [2, 5]
    cantidad = len(usuarios)

    if cantidad == 0:
        estado_ia = "MODO PASIVO: Esperando primer registro."
    elif cantidad < 3:
        estado_ia = "ESTABLE: Servidor con carga baja."
    elif cantidad < limite_capacidad:
        estado_ia = "PRECAUCIÓN: Carga de memoria aumentando."
    else:
        estado_ia = "CRÍTICO: Capacidad máxima. Bloqueando nuevos ingresos."

    # Interfaz de Usuario: Reporte de la IA y Menú
    print(f"\n[IA Reporta]: {estado_ia}")
    print(f"Usuarios en lista: {cantidad} | Espacio restante: {limite_capacidad - cantidad}")
    print("1. Registrar nuevo acceso")
    print("2. Consultar base de datos")
    print("3. Eliminar último registro")
    print("4. Apagar sistema")

    opcion = input("Seleccione una instrucción (1-4): ")

    # Lógica de decisión mediante if-elif-else [2, 6]
    if opcion == "1":
        # Uso de operadores relacionales para control de capacidad [6]
        if cantidad < limite_capacidad:
            nombre = input("Nombre del usuario a registrar: ")
            usuarios.append(nombre) # Añadir elemento al final de la lista [4]
            print(f">>> IA: Proceso exitoso. '{nombre}' tiene permiso de acceso.")
        else:
            print(">>> IA RECHAZA: Acceso denegado por falta de memoria en el servidor.")

    elif opcion == "2":
        if cantidad > 0:
            # Muestra la lista completa de datos almacenados [1]
            print(f">>> IA: Registros actuales en memoria: {usuarios}")
        else:
            print(">>> IA: No existen registros para mostrar.")

    elif opcion == "3":
        if cantidad > 0:
            # Elimina el último elemento de la lista para liberar espacio [5]
            eliminado = usuarios.pop()
            print(f">>> IA: Registro '{eliminado}' borrado correctamente.")
        else:
            print(">>> IA ERROR: La base de datos ya está vacía.")

    elif opcion == "4":
        # Condición para romper el ciclo while [3]
        sistema_activo = False
        print(">>> IA: Guardando registros y cerrando protocolos de seguridad...")

    else:
        print(">>> IA: Comando no reconocido. Por favor, repita la instrucción.")

print("\n--- SISTEMA FUERA DE LÍNEA ---")
```
